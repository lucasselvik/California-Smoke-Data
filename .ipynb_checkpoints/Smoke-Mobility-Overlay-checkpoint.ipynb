```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.animation import FuncAnimation
import json
```

Import data:

```python
smoke_csv = "./10km_grid/smokePM2pt5_predictions_daily_10km_20060101-20201231.csv"
smoke = pd.read_csv(smoke_csv)
smoke['date'] = pd.to_datetime(smoke['date'], format="%Y%m%d")
```

```python
mobility_csv = "california_202008_202009.csv"
mobility_data = pd.read_csv(mobility_csv)
```

```python
mobility_data.head()
```

```python
mobility_data['STOPS_LIST'] = mobility_data['STOPS_BY_DAY'].apply(json.loads)
mobility = mobility_data[['AREA', 'STOPS_LIST']].explode('STOPS_LIST')
mobility['Day'] = mobility.groupby('AREA').cumcount() + 1
mobility['Stops'] = mobility['STOPS_LIST'].astype(int)
mobility.head()
```

```python
stops_mean = np.mean(mobility['Stops'])
stops_sd = np.std(mobility['Stops'])
mobility['Stops_standardized'] = (mobility['Stops'] - stops_mean) / stops_sd
mobility.head()
```
